```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
import scipy as sc 
import statistics as st
```

```python
N = 10001

x = np.random.randn(N)
d1 = x-1
d2 = 2*x
d3 = x+1

sns.kdeplot(d1 , color='red')
sns.kdeplot(d2 , color='blue')
sns.kdeplot(d3 , color='green')
```

```python
## overlay the mean

# compute the means
mean_d1 = sum(d1) / len(d1)
mean_d2 = np.mean(d2)
mean_d3 = np.mean(d3)

plt.figure(figsize=(20,5))
# plot them
sns.kdeplot(d1 , color='red')
sns.kdeplot(d2 , color='blue')
sns.kdeplot(d3 , color='green')
plt.axvline(mean_d1, color='r', linestyle='--')
plt.axvline(mean_d2, color='b', linestyle='--')
plt.axvline(mean_d3, color='g', linestyle='--')

plt.xlabel('Data values')
plt.ylabel('Data counts')
```

```python
## "failure" of the mean

# new dataset of distribution combinations
d4 = np.hstack( (np.random.randn(N)-2,np.random.randn(N)+2) )
sns.kdeplot(d4)
mean_d4 = np.mean(d4)
plt.axvline(mean_d4 , color='b', linestyle='--')
plt.xlabel('Data values')
plt.ylabel('Data counts')
```

```python
## median

# create a log-normal distribution
shift   = 0
stretch = 0.7
n       = 2000

# generate data
data = stretch*np.random.randn(n) + shift
data = np.exp( data )

# compute mean and median
datamean = np.mean(data)
datamedian = np.median(data)

plt.figure(figsize=(20,5))
sns.kdeplot(data)
plt.axvline(datamean , color='g' ,  linestyle='--')
plt.axvline(datamedian , color='r' ,  linestyle='--')
```
