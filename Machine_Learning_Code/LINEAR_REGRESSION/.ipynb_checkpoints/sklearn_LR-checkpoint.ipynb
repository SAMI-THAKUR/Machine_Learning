```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url="https://drive.google.com/file/d/1VD06DjyGegNAWdJxFqKW-BtNsSbZsbez/view?usp=drive_link"
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
ad = pd.read_csv(url)
ad
```

<h2><mark>Visulaizing Data</mark></h2>

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 5))

sns.scatterplot(data=ad, x="TV", y='sales', ax=axes[0] , color='red')
sns.scatterplot(data=ad, x="radio", y='sales', ax=axes[1])
sns.scatterplot(data=ad, x="newspaper", y='sales', ax=axes[2] , color='green')
```

<h2><mark>Train Test Split</mark></h2>

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad.drop('sales' , axis=1)
y = ad['sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101) 
# test_size is percentage of test_data
# to compare algo thier random state must be same i.e Shuffle dataset
```

<h2><mark>Model Training</mark></h2>

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train , y_train)
```

```python
test_predictions = model.predict(X_test)
test_predictions
```

<h2><mark>Error Calculation</mark></h2>

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions) # AVERAGE #
MSE = mean_squared_error(y_test,test_predictions) # VARIANCE #
RMSE = np.sqrt(MSE) # Standard Deviation #
print("Mean Absolute Error:" , MAE ," Mean Square Error:",MSE , " Root Mean Square Error:",RMSE)
```

Comparing these metrics:

- **MAE** is the easiest to understand, because it's the average error.
- **MSE** is more popular than MAE, because MSE "punishes" larger errors, which tends to be useful in the real world.
- **RMSE** is even more popular than MSE, because RMSE is interpretable in the "y" units.

All of these are **loss functions**, because we want to minimize them.

<h2><mark>Linear Regression Eqation from Scikit</mark></h2>

```python
coeff = model.coef_
intercept = model.intercept_
print("Coefficient:" , coeff , " intercept" , intercept)
```

Interpreting the coefficients:
* Holding all other features fixed, a 1 unit (A thousand dollars) increase in TV Spend is associated with an increase in sales of  0.045 "sales units", in this case 1000s of units . 
* This basically means that for every $1000 dollars spend on TV Ads, we could expect 45 more units sold.---

* Holding all other features fixed, a 1 unit (A thousand dollars) increase in Radio Spend is associated with an increase in sales of  0.188 "sales units", in this case 1000s of units . 
* This basically means that for every $1000 dollars spend on Radio Ads, we could expect 188 more units sold.

* Holding all other features fixed, a 1 unit (A thousand dollars) increase in Newspaper Spend is associated with a **decrease** in sales of  0.001 "sales units", in this case 1000s of units . 
* This basically means that for every $1000 dollars spend on Newspaper Ads, we could actually expect to sell 1 less unit. Being so close to 0, this heavily implies that newspaper spend has no real effect on sales.

```python
# Compute predicted values
ad["predicted_values_y"] = intercept + coeff[0]*(ad['TV']) + coeff[1]*(ad['radio']) + coeff[2]*(ad['newspaper'])
ad.loc[:,("predicted_values_y","sales")]
```

<h2><mark>Residual Plotting</mark></h2>
It's also important to plot out residuals and check for normal distribution, this helps us understand if Linear Regression was a valid model choice.
<br/>
The more plot lie near zero the better the model is because it will be then follow normal distribution

```python
test_res = y_test - test_predictions
```

```python
sns.scatterplot(x=y_test,y=test_res)
plt.axhline(y=0, color='r', linestyle='--')
```

```python
sns.displot(data=test_res , bins=25,kde=True)
# sns.histplot(data=test_res ,bins=25 , kde=True) --> same
```
