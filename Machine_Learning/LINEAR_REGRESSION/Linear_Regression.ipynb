```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
url="https://drive.google.com/file/d/1VD06DjyGegNAWdJxFqKW-BtNsSbZsbez/view?usp=drive_link"
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
ad = pd.read_csv(url)
```

```python
X = ad.loc[: , 'TV':'newspaper']
y = ad['sales']
```

<h1><mark>MODEL TRAINING</mark></h1>

```python
X_test , X_train , y_test , y_train = train_test_split(X,y , random_state=101 , test_size=0.33)
```

```python
model = LinearRegression()
model.fit(X_test, y_test)
test_predictions = model.predict(X)
```

<h1><mark>RESIDUAL PLOT</mark></h1>

```python
residual = y - test_predictions
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))
sns.histplot(data=residual , bins=25 , kde=True , ax=axes[0])
sns.scatterplot(x=y,y=residual)
plt.axhline(y=0, color='r', linestyle='--')
```

<h1><mark>REGRESSION EQUATION</mark></h1>

```python
coefficients = model.coef_
# Get intercept
intercept = model.intercept_
# Print equation
print("Regression Equation:")
print("y = {:.4f} + {:.4f} * x1 + {:.4f} * x2^2 + {:.4f} * x3^3".format(intercept, coefficients[0], coefficients[1], coefficients[2]))
```
