```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
plt.plot(x, x**2, label="x**2");
plt.plot(x, x**3, label="x**3");
plt.legend(loc=0);

# ax.legend(loc=1) # upper right corner
# ax.legend(loc=2) # upper left corner
# ax.legend(loc=3) # lower left corner
# ax.legend(loc=4) # lower right corner
```

```python
plt.plot(x, x**2, 'b.-') # blue line with dots
plt.plot(x, x**3, 'v--') # green dashed line
```

```python
plt.plot(x, x+1, color="blue", alpha=0.5 , ls=":") # half-transparant
plt.plot(x, x+2, color="#8B008B" , ls="--")        # RGB hex code
plt.plot(x, x+3, color="#FF8C00", ls="-.")        # RGB hex code 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use marker for string code
# Use markersize or ms for size

ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=20)
```
