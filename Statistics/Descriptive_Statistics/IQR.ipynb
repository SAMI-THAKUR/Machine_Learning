```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Calculate the quartiles
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

# Create histogram
sns.histplot(data, kde=True , bins=10)

# Find the maximum height of the histogram bars
hist, bins = np.histogram(data, bins=10)
max_hist_height = np.max(hist)

# Overlay IQR region and fill between histogram and IQR region
plt.axvline(q1, color='green', linestyle='--', label=f'Q1: {q1:.2f}')
plt.axvline(q3, color='green', linestyle='--', label=f'Q3: {q3:.2f}')
plt.fill_betweenx([0, max_hist_height], q1, q3, color='green', alpha=0.3, label='IQR Region')

# Add legend and labels
plt.legend()
plt.title('Histogram with IQR Overlay')
plt.xlabel('Data')
plt.ylabel('Frequency')

```
