```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
```

```python
## example using log-normal distribution

# variable to evaluate the functions on
x = np.linspace(0,5,1001) # Return evenly spaced numbers over a specified interval.

# note the function call pattern...
p1 = stats.lognorm.pdf(x,1)
c1 = stats.lognorm.cdf(x,1)

p2 = stats.lognorm.pdf(x,.1)
c2 = stats.lognorm.cdf(x,.1)
```

```python
# draw the pdfs
fig,ax = plt.subplots(2,1,figsize=(4,7))
ax[0].plot(x,p1/sum(p1), x,p2/sum(p2))
ax[0].set_ylabel('probability')
ax[0].set_title('pdf(x)')

# draw the cdfs
ax[1].plot(x,c1, x,c2)
ax[1].set_ylabel('probability')
ax[1].set_title('cdf(x)')
plt.show()
```

```python
## computing the cdf from the pdf

# compute the cdf
c1x = np.cumsum(p1/sum(p1)) # cumulatiing and scaling b/w 0-1
c2x = np.cumsum(p2/sum(p2))

plt.plot(x,c1, x,c2)
plt.plot(x,c1x,'--')
plt.plot(x,c2x,'--')
plt.show()
```

```python
# note the function call pattern...
p1 = stats.t.pdf(x,1)
c1 = stats.t.cdf(x,1)

p2 = stats.t.pdf(x,.1)
c2 = stats.t.cdf(x,.1)
```

```python
## computing the cdf from the pdf

# compute the cdf
c1x = np.cumsum(p1/sum(p1)) # cumulatiing and scaling b/w 0-1
c2x = np.cumsum(p2/sum(p2))

plt.plot(x,c1, x,c2)
plt.show()
```

```python
# Define parameters of the normal distribution
mu = 2.5  # Mean
sigma = 0.5  # Standard deviation

# Calculate the normal PDF using the formula
pdf = 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-(x - mu)**2 / (2*sigma**2))

# Plot the normal PDF
plt.plot(x, pdf, label='Normal PDF', color='blue')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Probability Density Function')
plt.legend()
plt.grid(True)
plt.show()
```
