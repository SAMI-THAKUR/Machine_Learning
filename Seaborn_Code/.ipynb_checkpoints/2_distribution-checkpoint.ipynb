```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dm_office_sales.csv")
data["work experience"].value_counts()
data
```

```python
# RugPlot : individual data points are depicted as small marks or "ticks" along the axis, resembling the fringes of a rug.
# Histogram : frequency distribution of a set of continuous or discrete data
# KDE : is a method for smoothing a histogram to obtain a continuous probability distribution
```

```python
plt.figure(figsize=(8,8))
sns.rugplot(x="salary" , data=data , height=0.5)
```

```python
sns.set(style="darkgrid")
sns.displot(x="work experience" , data=data ,bins=16 , kde=True , rug=True)
```

```python
sns.kdeplot(data=data , x="work experience" , clip=[0,100])
```

```python
sns.histplot(data=data , x="work experience")
```
