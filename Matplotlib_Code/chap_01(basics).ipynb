```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0,100 , 10)
y = 2*x
```

```python
plt.plot(x,y);
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Graph Title Here')
plt.xlim(0,70)
plt.savefig('example.png')
```
