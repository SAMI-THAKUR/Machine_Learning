```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/file/d/1VzAeJybW9OCaYaz706te9wHJM_6wvyt9/view?usp=drive_link'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
sales= pd.read_csv(url)
sales
```

```python
plt.figure(figsize=(10,5), dpi=100)
sns.histplot(data=sales, x='sales', hue='level of education', multiple='stack', bins=20)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Sales by Level of Education')
plt.show()
```
