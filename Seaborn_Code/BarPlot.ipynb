```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/file/d/1VzAeJybW9OCaYaz706te9wHJM_6wvyt9/view?usp=drive_link'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
sales= pd.read_csv(url)
sales
```

<h2><mark>Simple CountPlot</mark></h2>

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=sales , x='division')
```

```python
plt.figure(figsize=(20,4),dpi=100)
sns.countplot(data=sales , x='level of education')
```

<h2><mark>Breakdown with hue</mark></h2>

```python
plt.figure(figsize=(20,5) , dpi=100)
sns.countplot(data=sales , x='level of education' , hue='training level')
```

<h2><mark>BarPlot</mark></h2>

<b><mark>so the bar plot below shows the mean salary on Y and education on X and the black line shows S.D i.e master degree has least standard deviation</mark></b><br><p>Not a good way of demonstating it</p>

```python
plt.figure(figsize=(10,5) , dpi=100)
sns.barplot(x='level of education' , y='sales' , data=sales , estimator=np.mean,errorbar='sd')
```
