```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
```

<h3><mark>Gaussian Distribution</mark></h3>

<h3><mark>Normal Distribution</mark></h3>

probability distribution that is symmetric around its mean, where the data tends to cluster near the mean with fewer observations further away from the mean


```python
## Gaussian Distribution ('Bell Curve') ## 
N = 1001
x = np.linspace(-5,5, N) # N evenly spaced numbers over the interval from -5 to 5 inclusive
gaussian_dist = stats.norm.pdf(x) # probability density function (pdf)

sns.lineplot(y=gaussian_dist/sum(gaussian_dist), x=x) 
plt.title('Gaussian Distribution')

print(sum(gaussian_dist)) # not a probability distribution becauuse sum is not 1
```

```python
## Normal Distribution ##
# parameters
stretch = 1 # variance (square of standard deviation)
shift   = 5 # mean
n       = 1000

data = stretch*np.random.randn(n,1) + shift
print(np.mean(data))

sns.histplot(data, kde=True)
plt.title('Empirical normal distribution')
plt.show()
```

<h3><mark>Uniformly Distribution</mark></h3>

```python
## Uniformly-distributed numbers ##

# parameters
stretch = 2 # not the variance
shift   = .5
n       = 10000

data = stretch*np.random.rand(n) + shift-stretch/2

fig,ax = plt.subplots(2,1,figsize=(5,6))

ax[0].plot(data,'.',markersize=1)
ax[0].set_title('Uniform data values')

sns.histplot(data , kde=True , ax=ax[1])
ax[1].set_title('Uniform data histogram')
```

```python
## binomial

# a binomial distribution is the probability of K heads in N coin tosses,
# given a probability of p heads (e.g., .5 is a fair coin).

n = 10 # number on coin tosses
p = .5 # probability of heads

x = range(n+2)
bindist = stats.binom.pmf(x,n,p)

plt.bar(x,bindist)
plt.title('Binomial distribution (n=%s, p=%g)'%(n,p))
```

```python
## t

x  = np.linspace(-4,4,1001)
df = 200
t  = stats.t.pdf(x,df)

plt.plot(x,t)
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) distribution'%df)
```

```python
## F

# parameters
num_df = 5   # numerator degrees of freedom
den_df = 100 # denominator df

# values to evaluate 
x = np.linspace(0,10,10001)

# the distribution
fdist = stats.f.pdf(x,num_df,den_df)

plt.plot(x,fdist)
plt.title(f'F({num_df},{den_df}) distribution')
plt.xlabel('F value')
```
