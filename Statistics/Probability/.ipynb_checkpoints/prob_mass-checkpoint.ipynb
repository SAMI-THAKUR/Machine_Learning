```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## compute empirical probability function

# continous signal (technically discrete!)
N = 10004
datats1 = np.cumsum(np.random.randn(N)) # Return the cumulative sum of the elements along a given axis.
datats2 = np.cumsum(np.random.randn(N)) 

# let's see what they look like
plt.plot(np.arange(N),datats1,linewidth=2)
plt.plot(np.arange(N),datats2,linewidth=2)
plt.show()


# discretize using histograms
nbins = 50

y,x = np.histogram(datats1,nbins) # y: height of bin
                                  # x: edges of bins
x1 = (x[1:]+x[:-1])/2 # center of each bins
y1 = y/sum(y)

y,x = np.histogram(datats2,nbins)
x2 = (x[1:]+x[:-1])/2
y2 = y/sum(y)


plt.plot(x1,y1, x2,y2,linewidth=3)
plt.legend(('ts1','ts2'))
plt.xlabel('Data value')
plt.ylabel('Probability')
plt.show()
```
