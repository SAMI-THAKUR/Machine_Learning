```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moment

# Generate some sample data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate the first four statistical moments
moment_1 = moment(data, moment=1)
moment_2 = moment(data, moment=2)
moment_3 = moment(data, moment=3)
moment_4 = moment(data, moment=4)


# Visualize the distribution
plt.figure(figsize=(15,10))
plt.hist(data, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram of Sample Data')
plt.xlabel('Data Value')
plt.ylabel('Density')

# Add vertical lines for the mean, standard deviation, skewness, and kurtosis
plt.axvline(moment_1, color='r', linestyle='--', label=f'Mean: {moment_1:.2f}')
plt.axvline(moment_2+moment_1, color='g', linestyle='--', label=f'Std Dev: {np.sqrt(moment_2):.2f}')
plt.axvline(moment_1-moment_2, color='g', linestyle='--', label=f'Std Dev: {np.sqrt(moment_2):.2f}')
plt.fill_betweenx([0, 1] , moment_2+moment_1,moment_1-moment_2,color='g', alpha=0.3)
plt.axvline(moment_3, color='b', linestyle='--', label=f'Skewness: {moment_3:.2f}')
plt.axvline(moment_4, color='purple', linestyle='--', label=f'Kurtosis: {moment_4:.2f}')

plt.legend()

```
